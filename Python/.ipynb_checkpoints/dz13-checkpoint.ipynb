```python
# как разбирать всю эту разметку? Поможет BeautifulSoup.
from bs4 import BeautifulSoup
```

```python
# определяем список ключевых слов, которые нам интересны
# добавила c++ так как в свежих постах не упомнилася парсинг и python
KEYWORDS = ['python', 'парсинг', 'c++']
```

```python
# получаем страницу с самыми свежими постами
req = requests.get('https://habr.com/ru/all/')
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
# извлекаем посты
posts = soup.find_all('article', class_='post')
posts
```

```python
for post in posts:
    post_id = post.parent.attrs.get('id')
   # если идентификатор не найден, это что-то странное, пропускаем
    if not post_id:
        continue
    post_id = int(post_id.split('_')[-1])
    #print('post', post_id)
    previews = post.find_all('div', class_='post__text')
    print('preview', previews)
```

```python
# добавляем извлечение ключевых слов из постов, чтобы отбирать только нужные
for post in posts:
    post_id = post.parent.attrs.get('id')
   # если идентификатор не найден, это что-то странное, пропускаем
    if not post_id:
        continue
    post_id = int(post_id.split('_')[-1])
    previews = post.find_all('div', class_='post__text')
    for preview in previews:
            preview_lower = preview.text.lower()
            #print(preview_lower)
           # ищем вхождение хотя бы одного желаемого хаба
            if any([desired in preview_lower for desired in KEYWORDS]):
               # пост нам интересен - делаем с ним все что захотим:
               # можно отправит в телеграм уведомление, можно на почту и т.п.
                #print(preview_lower)
                title_element = post.find('a', class_='post__title_link')
                print(title_element.text, title_element.attrs.get('href'))
               # так как пост уже нам подошел - дальше нет смысла проверять хабы
                #break
```

```python
import pandas as pd
```

```python
# запишем данные в DataFrame
df = pd.DataFrame()
for post in posts:
    post_id = post.parent.attrs.get('id')
   # если идентификатор не найден, это что-то странное, пропускаем
    if not post_id:
        continue
    post_id = int(post_id.split('_')[-1])
    previews = post.find_all('div', class_='post__text')
    for preview in previews:
            preview_lower = preview.text.lower()
            #print(preview_lower)
           # ищем вхождение хотя бы одного желаемого хаба
            if any([desired in preview_lower for desired in KEYWORDS]):
               # пост нам интересен - делаем с ним все что захотим:
               # можно отправит в телеграм уведомление, можно на почту и т.п.
                data_element = post.find('span', class_='post__time')
                data = data_element.text
                title_element = post.find('a', class_='post__title_link')
                title = title_element.text
                url = title_element.attrs.get('href')
                row = {'data': data, 'title': title, 'Url': url}
                df = pd.concat([df, pd.DataFrame([row])])
df.reset_index()
```
