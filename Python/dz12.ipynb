```python
rating = pd.read_csv('part education/доп_материалы_к_пандас_1_и_2/ml-latest-small/ratings.csv')
rating.head()
```

```python
rating_count=rating.groupby('userId').count().reset_index()
rating_count = rating_count[['userId', 'rating']]
rating_count.head()
```

```python
user_a_lot_of_rating = rating_count[rating_count.rating > 99]
user_a_lot_of_rating.head()
```

```python
timestamp_max = rating.groupby('userId').max().reset_index()
timestamp_max = timestamp_max[['userId', 'timestamp']]
timestamp_min = rating.groupby('userId').min().reset_index()
timestamp_min = timestamp_min[['userId', 'timestamp']]
```

```python
timestamp_max.rename(columns={'timestamp': 'timestamp_max'}, inplace=True)
timestamp_min.rename(columns={'timestamp': 'timestamp_min'}, inplace=True)
```

```python
joined = user_a_lot_of_rating.merge(timestamp_max, on='userId', how='left')
joined1 = joined.merge(timestamp_min, on='userId', how='left')
joined1.head(5)
```

```python
joined1['liveTime'] = joined1['timestamp_max'] - joined1['timestamp_min']
joined1.head()
```

Дана статистика услуг перевозок клиентов компании по типам:

- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
table_lite = rzd.merge(auto, how='outer', on='client_id')
table_lite.head()
```

```python
table_lite_all = table_lite.merge(air, how='outer', on='client_id')
table_lite_all.head()
```

```python
table_all_inf = table_lite_all.merge(client_base, how='outer', on='client_id')
table_all_inf.head(10)
```
