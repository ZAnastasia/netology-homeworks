```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np
import scipy.stats #при работе со статистикой, там большое количество статистических тестов
```

```python
Выбрала особенности:
● Если мы увеличиваем, стоимость поездки в часы повышенного спроса, 
то для устройств Android наблюдаем более высокие юзерские отмены, 
что може приводит к потере дохода
● Если мы увеличиваем, стоимость поездки в часы повышенного спроса, 
то в Заказах из центра, процент юзерских отмен значительно ниже,
чем из других районов города, соответвено можем раскатить фичу на заказы из центра 
и получить повышение прибыли    
```

```python
df = pd.read_csv('Netology_A_B-testing_HW_1.csv')
df.head()
```

```python
df_n = df[['date','user_id','hour','app_opened','price_seen','order_made', 'surge', 'ride_completed', 'user_cancelled', 'age', 'os', 'city_center_order', 'order_class', 'distance', 'rfm']]
df_n.head()
```

```python
df_n.head()
```

```python
df_n.info()
```

```python
df_new = df_n.dropna()
df_new.info()
```

```python
df_new.head()
```

```python
#добавляем новую колонку, с рандомным разделение мна группы 1 и 2
df_new['randNumGroup'] = np.random.randint(1,3, size=len(df_new))
df_new.head(7)
```

```python
#фильтр по группам
filter1_df_new = df_new[(df_new['randNumGroup'] == 1)] 
filter2_df_new = df_new[(df_new['randNumGroup'] == 2)] 
filter1_df_new.head()
```

```python
filter2_df_new.head()
```

```python
filterdf_grou1 = filter1_df_new[['user_id','app_opened','price_seen','order_made', 'surge', 'ride_completed', 'user_cancelled', 'os']]
filterdf_grou1.head()
```

```python
filterdf_grou2 = filter2_df_new[['user_id','app_opened','price_seen','order_made', 'surge', 'ride_completed', 'user_cancelled', 'os']]
filterdf_grou2.head()
```

```python
#сравниваем размер групп
print(len(filterdf_grou1['os']))
print(len(filterdf_grou2['os']))
```

```python
filterdf_grou2 = filterdf_grou2.iloc[:45672]
```

```python
#выравниваем рамер групп
print(len(filterdf_grou1['os']))
print(len(filterdf_grou2['os']))
```

```python
from scipy import stats
t,p = stats.ttest_ind(filterdf_grou1['user_cancelled'],filterdf_grou2['user_cancelled'])
t,p
```
